```python
df_2017 = pd.merge(meta, credits, on='id')
```

```python
import ast
df_2017['genres'] = df_2017['genres'].map(lambda x: ast.literal_eval(x))
df_2017['cast'] = df_2017['cast'].map(lambda x: ast.literal_eval(x))
df_2017['crew'] = df_2017['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def make_genresList(x):
    gen = []
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (" ".join(gen))
```

```python
df_2017['genres'] = df_2017['genres'].map(lambda x: make_genresList(x))
```

```python
def get_actor(x , n):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    if len(casts)<=n:
        return np.NaN
    else:
        return (casts[n])
```

```python
df_2017['actor_1_name'] = df_2017['cast'].map(lambda x: get_actor(x,0))
```

```python
df_2017['actor_1_name']
```

```python
df_2017['actor_2_name'] = df_2017['cast'].map(lambda x: get_actor(x,1))
```

```python
df_2017['actor_2_name']
```

```python
df_2017['actor_3_name'] = df_2017['cast'].map(lambda x: get_actor(x,2))
```

```python
df_2017['actor_3_name']
```

```python
df_2017['crew'][0]
```

```python
def get_directors(x):
    dt = []
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (" ".join(dt))
```

```python
df_2017['director_name'] = df_2017['crew'].map(lambda x: get_directors(x))
```

```python
df_2017['director_name']
```

```python
df_2017 = df_2017[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','title']]
```
